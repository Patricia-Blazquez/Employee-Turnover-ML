#### **EDA - CATEGORICAL VARIABLES** (VARIABLES CATEGÓRICAS)

```python
def graficos_eda_categoricos(cat):

    #Calculate the number of rows we need / Calcular el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    #Define the plot / Definir el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Flatten to iterate through the plot as if it were 1-dimensional instead of 2/ Aplanar para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat

    #Create the loop that adds the plots / Crear el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(df.select_dtypes('O'))
```

  * **satisfaccion_trabajo (job satisfaction)**: Impute to 'Alta'(High)

**Conclusiones**:

### **EDA - NUMERIC VARIABLES** (VARIABLES NUMÉRICAS)

```python
def estadisticos_cont(num):
    #Calculate describe / Calcular describe
    estadisticos = num.describe().T
    #Add the median / Añadir la mediana
    estadisticos['median'] = num.median()
    #Reorder to place the median next to the mean / Reordenar para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Returns it / Lo devuelve
    return(estadisticos)
```

```python
estadisticos_cont(df.select_dtypes('number'))
```

### What is the economic impact of this problem?

¿Cual es el impacto económico de este problema?

According to the "**Cost of Turnover**" study by the Center for American Progress:

* The **cost** of turnover for employees earning **less than USD 30,000** is **16.1%** of their salary
* The **cost** of turnover for employees earning **between USD 30,000-50,000** is **19.7%**of their salary
* The **cost** of turnover for employees earning **between USD 50,000-75,000** is **20.4%** of their salary
* The **cost** of turnover for employees earning **more than USD 75,000** is **21%** of their salary



Según el estudio "**Cost of Turnover**" del Center for American Progress:

* El coste de la fuga de los empleados que ganan menos de 30000 es del 16,1% de su salario

* El coste de la fuga de los empleados que ganan entre 30000-50000 es del 19,7% de su salario

* El coste de la fuga de los empleados que ganan entre 50000-75000 es del 20,4% de su salario

* El coste de la fuga de los empleados que ganan más de 75000 es del 21% de su salario

```python
# Create a new variable: employee_annual_salary / Crear nueva variable salario_ano del empleado
df['salario_ano'] = df.salario_mes.transform(lambda x: x*12)
df[['salario_mes','salario_ano']]
```

```python
#Calculate the economic impact of each employee if they leave the company / Calcular el impacto económico de cada empleado si deja la empresa

#List of conditions / Lista de condiciones
condiciones = [(df['salario_ano'] <= 30000),
               (df['salario_ano'] > 30000) & (df['salario_ano'] <= 50000),
               (df['salario_ano'] > 50000) & (df['salario_ano'] <= 75000),
               (df['salario_ano'] > 75000)]

#List of results / Lista de resultados
resultados = [df.salario_ano * 0.161, df.salario_ano * 0.197, df.salario_ano * 0.204, df.salario_ano * 0.21]

#Apply select / Aplicar select
df['impacto_abandono'] = np.select(condiciones,resultados, default = -999)

df
```

### We can continue developing strategies based on turnover insights:
It was seen that sales representatives are the position with the highest turnover. Would it make sense to create a specific plan for them? What would be the cost savings if we reduce turnover by 30%?

First, calculate the % of sales representatives who left last year.


**Se puede seguir trazando estrategias asociadas a los insights de abandono:**

Se vio que los representantes de ventas son el puesto que más se van. ¿Tendría sentido hacer un plan específico para ellos? ¿Cual sería el coste ahorrado si disminuimos la fuga un 30%?

Primero calcular el % de representantes de ventas que se han ido el año pasado

```python
total_repre_pasado = len(df.loc[df.puesto == 'Sales Representative'])
abandonos_repre_pasado = len(df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 1)])
porc_pasado = abandonos_repre_pasado / total_repre_pasado

porc_pasado
```

Now estimate how many will leave this year (Ahora estimar cuántos se irán este año)

```python
total_repre_actual = len(df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 0)])
se_iran = int(total_repre_actual * porc_pasado)

se_iran
```

Of those, **how many** can be retained (hypothesis 30%) and **how much** money could this save.

(Sobre ellos, cuántos se pueden retener (hipótesis 30%) y cuánto dinero puede suponer)

```python
retenemos = int(se_iran * 0.3)

ahorramos = df.loc[(df.puesto == 'Sales Representative') & (df.abandono == 0),'impacto_abandono'].sum() * porc_pasado * 0.3

print(f'Podemos retener {retenemos} representantes de ventas y ello supondría ahorrar {ahorramos}$.')
```

## MACHINE LEARNING MODEL (MODELO DE MACHINE LEARNING)

```python
df_ml = df.copy()
```

```python
df_ml.info()
```

### DATA PREPARATION FOR MODELING (PREPARACIÓN DE LOS DATOS PARA LA MODELIZACIÓN)

#### Transform all categorical variables to numeric
(Transformar todas las variables categóricas a númericas)

```python
from sklearn.preprocessing import OneHotEncoder

#Categorical / Categóricas
cat = df_ml.select_dtypes('O')

#Instantiate / Instanciar
ohe = OneHotEncoder(sparse_output = False)

#Train / Entrenar
ohe.fit(cat)

#Apply / Aplicar
cat_ohe = ohe.transform(cat)

#Name the variables / Poner los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```

```python
cat_ohe
```

#### Final DataFrame (Dataframe final)

* Select the numeric variables to join them with the cat_ohe (Seleccionar las variables numéricas para poder juntarlas a las cat_hoe)

```python
num = df.select_dtypes('number').reset_index(drop = True)
```

* Merge all into the final dataframe (Juntar todas en el dataframe final)

```python
df_ml = pd.concat([cat_ohe,num], axis = 1)
df_ml
```

### MODELING DESIGN (DISEÑO DE LA MODELIZACIÓN)

#### Separation of predictors and target (Separación predictoras y target)

```python
x = df_ml.drop(columns='abandono')
y = df_ml['abandono']
```

#### Train-test split (Separación train y test)

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3)
```

### MODEL TRAINING ON TRAIN (ENTRENAMIENTO DEL MODELO SOBRE TRAIN)

```python
from sklearn.tree import DecisionTreeClassifier

# Instantiate / Instanciar
ac = DecisionTreeClassifier(max_depth=4)

# Train / Entrenar
ac.fit(train_x,train_y)
```

### PREDICTION AND VALIDATION ON TEST (PREDICCIÓN Y VALIDACIÓN SOBRE TEST)

```python
# Prediction / Predicción
pred = ac.predict_proba(test_x)[:, 1]
pred[:20]
```

```python
# Evaluation / Evaluación
from sklearn.metrics import roc_auc_score

roc_auc_score(test_y,pred)
```

### INTERPRETATION (INTERPRETACIÓN)

#### Tree diagram (Diagrama del árbol)

```python
from sklearn.tree import plot_tree

plt.figure(figsize = (50,50))

plot_tree(ac,
          feature_names= test_x.columns,
          impurity = False,
          node_ids = True,
          proportion = True,
          rounded = True,
          precision = 2);
```

#### Variable importance (Importancia de las variables)

```python
pd.Series(ac.feature_importances_,index = test_x.columns).sort_values(ascending = False).plot(kind = 'bar', figsize = (30,20));
```

### DEPLOYMENT (EXPLOTACIÓN)

Incorporate scoring into the main dataframe (Incoporación del scoring al dataframe principal)

```python
df['scoring_abandono'] = ac.predict_proba(df_ml.drop(columns = 'abandono'))[:, 1]
df
```
